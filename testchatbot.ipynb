```python
import pyocr
from PIL import Image, ImageGrab
import pyperclip as pc
import pyautogui
import pyautogui as pag
import time
import pyperclip
import datetime
import random
import schedule

pyocr.tesseract.TESSERACT_CMD = r'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'

tools = pyocr.get_available_tools()
tool = tools[0]

#座標は全て0に指定しています　各自座標を指定してください
#time.sleep(N)で次の行動までのインターバルは各自指定してください　指定されてる時間はあくまで例です

def job():


    pag.click(x=0, y=0)
    time.sleep(0.1)

    # ()内の言葉をコピー

    pyperclip.copy("良しくお願いします。")
    time.sleep(0.1)
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('enter')
    time.sleep(0.5)
    
    #　クリックする座標
    pag.click(x=0, y=0)
        
    i = 0


#　ループする回数を指定

    while i < 100:
        
        pag.click(x=0, y=0)
        time.sleep(0.5)
        pyautogui.hotkey('win', 'shift', 's',  button='left')
        time.sleep(0.5)
        pyautogui.dragTo(0, 0, 0, button='left')    


        img = ImageGrab.grabclipboard()
        #img = Image.open("./test.png")

        builder = pyocr.builders.TextBuilder(tesseract_layout=6)
        text = tool.image_to_string(img, lang="PNG", builder=builder)

        pc.copy(text)


        string = text

#　チャットの語句を格納する配列

        mylist = ["何が食べたいですか？","もう少しだけ頑張りましょう" ,"休憩する時間はありませんよ"
                ]



        temp = ("お腹すいた" in string)
        temp1 = ("もうすぐ休憩だ" in string)
        temp2 = ("疲れた" in string)
        first = ("おはようございます" in string)
        second = ("こんにちは" in string)
        third = ("こんばんは" in string)

        

        time.sleep(2)
        


        if (temp or temp1 or temp2):
            word = (random.choice(mylist))

            pag.click(x=0, y=0)
            time.sleep(0.1)
            pyperclip.copy(word)
            time.sleep(0.1)
            pyautogui.hotkey('ctrl','v')
            pyautogui.press('enter')
            time.sleep(0.5)
            pag.click(x=0, y=0)
            time.sleep(0.5)
            pag.click(x=0, y=0)
            time.sleep(0.5)
            pag.click(x=0, y=0)
            pyperclip.copy("") 


        elif(first):
            test_ai("おはようございます","今日はまだ始まったばかりです","楽しみま")    
        


        elif(second):
            test_ai("こんにちは","今日もいい天気ですね","今日も１日頑張りましょう！")


        elif(third):

            test_ai("こんばんは","今日も１日お疲れ様です","帰ってゆっくり休んでください")


            time.sleep(4.5)

        i += 1

    else:
        pag.click(x=5573, y=1415)
        time.sleep(0.1)
        pyperclip.copy("お先に失礼します。")
        time.sleep(0.1)
        pyautogui.hotkey('ctrl','v')
        pyautogui.press('enter')
        time.sleep(0.5)
        pag.click(x=5246, y=1154)  
        pyperclip.copy("")    


def test_ai(word,word2,word3):


    pag.click(x=0, y=0)
    time.sleep(0.1)
    pyperclip.copy(word)
    time.sleep(0.1)
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('enter')
    time.sleep(0.5)
    pag.click(x=0, y=0)

    pag.click(x=0, y=0)
    time.sleep(0.1)
    pyperclip.copy(word2)
    time.sleep(0.1)
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('enter')
    time.sleep(0.5)
    pag.click(x=0, y=0)

    pag.click(x=5573, y=1415)
    time.sleep(0.1)
    pyperclip.copy(word3)
    time.sleep(0.1)
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('enter')
    time.sleep(0.5)
    pag.click(x=0, y=0)
    time.sleep(0.5)
    pag.click(x=0, y=0)    
    pyperclip.copy("")      



       
#プログラムを実行する時間を指定
schedule.every().day.at("15:00").do(job)

    
while True:
  schedule.run_pending()
  time.sleep(10)
```
